##  Визуальное исследование датасета

Для визуального исследования могут быть использованы различные виды диаграмм, мы построим только некоторые варианты диаграмм, которые используются достаточно часто.

### Диаграммма рассеяния

```python
fig, ax = plt.subplots(figsize=(10,10)) 
sns.scatterplot(ax=ax, x='residual sugar', y='density', data=data)
```

```python
fig, ax = plt.subplots(figsize=(10,10)) 
sns.scatterplot(ax=ax, x='residual sugar', y='density', data=data, hue='quality')
```

```python
fig, ax = plt.subplots(figsize=(10,10)) 
sns.distplot(data['residual sugar'])
```

### Jointplot

```python
sns.jointplot(x='free sulfur dioxide', y='total sulfur dioxide', data=data)
```

```python
sns.jointplot(x='free sulfur dioxide', y='total sulfur dioxide', data=data, kind="hex")
```

```python
sns.jointplot(x='free sulfur dioxide', y='total sulfur dioxide', data=data, kind="kde")
```

## Информация о корреляции признаков

```python
data.corr()
```

```python
data.corr(method='pearson')
```

```python
data.corr(method='spearman')
```

### Тепловая карта

```python
sns.heatmap(data.corr())
```

```python
# Вывод значений в ячейках
sns.heatmap(data.corr(), annot=True, fmt='.3f')
```

```python
# Треугольный вариант матрицы
mask = np.zeros_like(data.corr(), dtype=np.bool)
# чтобы оставить нижнюю часть матрицы
# mask[np.triu_indices_from(mask)] = True
# чтобы оставить верхнюю часть матрицы
mask[np.tril_indices_from(mask)] = True
sns.heatmap(data.corr(), mask=mask, annot=True, fmt='.3f')
```

```python
fig, ax = plt.subplots(1, 3, sharex='col', sharey='row', figsize=(15,5))
sns.heatmap(data.corr(method='pearson'), ax=ax[0], annot=True, fmt='.2f')
sns.heatmap(data.corr(method='kendall'), ax=ax[1], annot=True, fmt='.2f')
sns.heatmap(data.corr(method='spearman'), ax=ax[2], annot=True, fmt='.2f')
fig.suptitle('Корреляционные матрицы, построенные различными методами')
ax[0].title.set_text('Pearson')
ax[1].title.set_text('Kendall')
ax[2].title.set_text('Spearman')
```
